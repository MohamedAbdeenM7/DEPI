#### Functions to string

```python
name = "mostafa abdeeN mostafa"
print(name.replace("mostafa","mohamed"))
print(name.capitalize())
print(name.title()) # and there is upper an lower 
print("# occurrences",name.count("mostafa"))
print(name.split())
print(name.swapcase()) #notice the n in abdeen
```
