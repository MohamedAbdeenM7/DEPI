### Problem 1 <b>(Reversed Words)</b>
##### Description:

Complete the solution so that it reverses all of the words within the string passed in.<br>
Words are separated by exactly one space and there are no leading or trailing spaces.<br>

Example(Input --> Output):<br>
"The greatest victory is that which requires no battle" --> "battle no requires which that is victory greatest The"

```python
# Task 1 Solution

sentance = ('The greatest victory is that which requires no battle')
words = list(sentance.split(" "))
re_words = words[::-1]
re_sentance = " "
for i in re_words:
   print(re_sentance + i)

```

### Problem 2 <b>(Are You Playing Banjo?)</b>
#### Description:

Create a function which answers the question "Are you playing banjo?".<br>
If your name starts with the letter "R" or lower case "r", you are playing banjo!<br>
The function takes a name as its only argument, and returns one of the following strings:<br>

name + " plays banjo"<br> 
name + " does not play banjo"<br>
Names given are always valid strings.<br>

```python
def banjio(name: str):
    if name and name[0].lower() == 'r':
        return f"{name}, you are playing Banjo"
    else:
        return f"{name}, you are not playing Banjo"


name = input("Please Enter Your Name: ")
print(banjio(name))
```
