```python
# Load my dataset using pandas Dataframe

data = pd.read_csv("ford_go_bike_data.csv")
df = data.copy()
```

```python
import pandas as pd


def overview_data(df):
    overview = pd.DataFrame(
        {
            # "Column": df.columns,
            "Type": df.dtypes,
            "Non-Null": df.count(),
            "Null": df.isnull().sum(),
            "Unique": df.nunique(),
            "Sample": df.iloc[0] if len(df) > 0 else "",
        }
    )

    print(f"Rows: {df.shape[0]} | Columns: {df.shape[1]}")
    print(f"Memory: {df.memory_usage().sum()/1024**2:.1f} MB")

    return overview
```

###  To fix this issue You will fill these values with mean for numeric and mod for categorical

```python
# trying to fill missing values in my datset using for loop
# we will delete the missing values for start_station_id , start_station_name , end_station_id , end_station_name
columns_to_fill = ['start_station_id', 'start_station_name', 'end_station_id', 'end_station_name']
for column in columns_to_fill:
    df[column].fillna('Unknown', inplace=True)

# for member_birth_year we will fill with mean value
df['member_birth_year'].fillna(df['member_birth_year'].mean(), inplace=True)
```

 > ## From all previous results above I found  :

> 1 ) **unique Value** as in ((`member_gender`)) : Not significant , so won't clean it



> 2 ) **Incorrect data type** as in (( `user_type` &&	`member_gender`)) :  significant change , so I need to clean it.

```python
##  step  2 ) Incorrect data type as in (( `user_type` &&	`member_gender`)) :  significant change , so  need to clean it.
df['user_type'] = df['user_type'].astype('category')
df['member_gender'] = df['member_gender'].astype('category')
```

```python
## check your work
df.info()
```

```python
## step 3 ) Good factor for marketing work as in ((`member_birth_year`)) : so need to clean it.
## To know the target audience's age and serve on them

# create a column for users' age depending on the current year
current_year = 2024
df['member_age'] = current_year - df['member_birth_year']
# check the new column
df[['member_birth_year', 'member_age']].head(10)
df[['member_birth_year','member_age']]
```

```python
## check your work
df['member_age'].head(10)
```

```python
## change the new column age data type from float to int
# round the age values to the nearest whole number
df['member_age'] = df['member_age'].round().astype('int')
## check your work
df['member_age'].head(10)
```

```python
## create a function to know the average users ages
def average_user_age(dataframe):
    return dataframe['member_age'].mean()

avg_user = average_user_age(df)
print("The average user age is:", avg_user.round(2))

```

```python
## another important factor to offer  marketing services is the  duration sec time 
## need to change the time from second to minute to be more usable
df['trip_duration_minutes'] = df['duration_sec'] / 60
## check your work
df['trip_duration_minutes'].head(10)
```

```python
## I need to change the time from second to Houre  to be more usable
df['trip_duration_hours'] = df['duration_sec'] / 3600
```

```python
## check your work
df['trip_duration_hours'].head(10)
```

# <a id="8"> 8) . What the average duration of trip for the categories of user type?</a>

```python
# the average duration of trip for the categories of user type
avg_duration_by_user_type = df.groupby('user_type')['trip_duration_minutes'].mean().reset_index()
plt.figure(figsize=(8, 6), dpi=200)
sb.barplot(data=avg_duration_by_user_type, x='user_type', y='trip_duration_minutes')
plt.xlabel('User Type')
plt.ylabel('Average Trip Duration (Minutes)')
plt.title('Average Trip Duration by User Type')
plt.show()
```
