```python
class Member:

    not_allowed_names = ["Hell","Shit","Baloot"]
    users_num = 0
    
    
    @classmethod
    def show_users_count(cls):
        return f"the count of our users in the system is {cls.users_num}"

    @staticmethod
    def say_hello():
        return f"Hello"


    def __init__(self,first_name,middle_name,last_name,gendar):
        self.fname = first_name
        self.mname = middle_name
        self.lname = last_name
        self.gendar = gendar
        Member.users_num += 1
        
    def full_name(self):

        if self.fname in Member.not_allowed_names:
            raise ValueError("name not allowed")
        return f"{self.fname} {self.mname} {self.lname}"

    def name_with_title(self):
        if self.gendar == "Male":
            return f"Hello Mr.{self.fname}"
        elif self.gendar == "Female":
            return f"Hello Miss.{self.fname}"
        else:
            return f"Hello {self.fname}"
    
    def get_all_info(self):
        return f"{self.name_with_title()} , Your Full Name Is: {self.full_name()}"
    
    def delete_user(self) :
        Member.users_num -= 1

        return f"User {self.fname} Deleted"

member_one = Member("Mohamed","Abdeen","Mostafa","Male")
member_two = Member("Mirall","Abdeen","Mostafa","Female")
member_three = Member("mostafa","Abdeen","Mostafa","bomba")
member_four = Member("Hell","kasa","delta","bomba")

# print(member_one.name_with_title()), print(member_two.name_with_title()), print(member_three.name_with_title())

print(member_one.get_all_info())
print(Member.users_num)
```
