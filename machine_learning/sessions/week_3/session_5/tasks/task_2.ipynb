#### Task 2 – <b>(Thanos Project)</b>  
Objective:<br>
Create a Python function that performs a series of file and folder operations,<br>
inspired by the 
concept of “Thanos” randomly removing half of the files.<br>
• The final number of files is displayed correctly. <br>

```python
import os
import random
def thanos_snap():
    os.makedirs("thanos", exist_ok=True)

    # Create 10 files
    for i in range(20):
        open(f"thanos/file{i}.txt", "w").close()

    items = os.listdir("thanos")
    half_items = len(items) // 2

    # Pick half of the files randomly (no duplicates)
    choices = random.sample(items, half_items)

    choices_paths = []
    for choice in choices:
        file_path = os.path.join("thanos", choice)
        choices_paths.append(file_path)
        os.remove(file_path)

    print("Deleted files:", choices)
```

```python
thanos_snap()
```
