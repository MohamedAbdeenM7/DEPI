#### Task 1 – <b>(Simple Calculator Program)</b>
Objective:<br>
Develop a Python program that acts as a simple calculator. The program should allow the user to<br>
choose an arithmetic operation and apply it to two numbers entered by the user.<br>

```python
def add(n1,n2):
    return n1 + n2
def sub(n1,n2):
    return n1 - n2
def mul(n1,n2):
    return n1 * n2
def div(n1,n2):
    return n1 / n2

def calc():
    print("Simple Calculator")
    n1 = float(input("Please Entere The First Number:"))
    operation = input("Please Enter The Desierd Operation:( +, -, *, / )")
    n2 = float(input("Please Entere The Second Number:"))

    if operation   == "+":
        result = add(n1,n2)
    elif operation == "-":
        result = sub(n1,n2)
    elif operation == "*":
        result = mul(n1,n2)
    elif operation == "/":
        result = div(n1,n2)
    else:
        print("error")
    return print(n1,operation,n2,"= ",result)
```

```python
calc()
```

#### Task 2 – <b>(Thanos Project)</b>  
Objective:<br>
Create a Python function that performs a series of file and folder operations,<br>
inspired by the 
concept of “Thanos” randomly removing half of the files.<br>
• The final number of files is displayed correctly. <br>

```python
import os
import random
def thanos_snap():
    os.makedirs("thanos", exist_ok=True)

    # Create 10 files
    for i in range(20):
        open(f"thanos/file{i}.txt", "w").close()

    items = os.listdir("thanos")
    half_items = len(items) // 2

    # Pick half of the files randomly (no duplicates)
    choices = random.sample(items, half_items)

    choices_paths = []
    for choice in choices:
        file_path = os.path.join("thanos", choice)
        choices_paths.append(file_path)
        os.remove(file_path)

    print("Deleted files:", choices)
```

```python
thanos_snap()
```
