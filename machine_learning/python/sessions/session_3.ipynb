```python
def sum(num_1,num_2):

    return float(num_1) + float(num_2)
def sub(num_1,num_2):

    return float(num_1) - float(num_2)
def mul(num_1,num_2):

    return float(num_1) * float(num_2)
def div(num_1,num_2):

    return float(num_1) / float(num_2)
def main():
    print("--------calc--------")
    print("choose operations")
    print("1-sum")
    print("2-sub")
    print("3-mul")
    print("4-div")

    choice = input("Enter The Opreation")
    num_1  = input("Enter Num_1")
    num_2  = input("Enter Num_2")

    if choice == "1":
        result = sum(num_1,num_2)
    elif choice == "2":
        result = sub(num_1,num_2)
    elif choice == "3":
        result = mul(num_1,num_2)
    elif choice == "4":
        result = div(num_1,num_2)
    else:
        print("error")
        
    return print(result)
```

```python
if __name__ == "__main__":
    main()
```
