```python
from abc import ABC, abstractmethod

class Animal(ABC):

    # @abstractmethod
    def move(self):
        pass

class Bird(Animal):

    def move(self):
        print("Move from Bird")

class Cat(Animal):
    
    def move(self):
        print("Move from Cat")
```

```python
from prettytable import PrettyTable
 
products = [{"name":"water","price":80.0,"quantity":1500},
            {"name":"Soda","price":130.0,"quantity":1500},
            {"name":"Chips","price":75.0,"quantity":1500},
            {"name":"Bread","price":45.0,"quantity":1500},
            {"name":"Eggs","price":200.0,"quantity":1500}
             ]
table = PrettyTable()
table.field_names = ["Product Name","Price($)", "Quantity"]
for p in products:
    table.add_row([p["name"],f"{p['price']:.2f}", p["quantity"]])
print(table)
total_bill = 0
 
while True:
    product_name = input("\nEnter product name (or 'exit') to finish: ").strip().lower()
    if product_name == 'exit':
        break
    product = next((p for  p in products if p['name'].lower() == product_name), None)
   
    if product is None:
        print("product not found. please try again")
        continue
    try:
        quantity_required = float(input(f"How many {product['name']} do you want?"))
        if quantity_required <=0:
            print("Quantity must be greater than Zero")
            continue
        if quantity_required > product['quantity']:
            print(f"Insufficient quantity!")
            continue
        discount_percentage = 0.05
        price_before_disc = quantity_required * product['price']
        discount_amount = price_before_disc * discount_percentage
        final_price = price_before_disc - discount_amount
       
        product['quantity'] -= quantity_required
        total_bill += final_price
        print(f"Success! price for {product['name']}: ${final_price:.2f}(saved: ${discount_amount:.2f})")
        another = input("something else? (y/n)").lower()
        if another != 'y':
            break
    except ValueError:
        print("Invalid input! pls enter a number for qty")
print(f"\n Total Amount to pay: ${total_bill:.2f}")
print("thank you !")
 
```

```python
def factorial(num):
    if num == 0 or num == 1:
        return 1
    return num * factorial(num-1) # multiply the num by num-1 but it will repeat the function untill num will hit 1 or 0

factorial(7)
```

```python
class Course:
    id_counter = 1

    def __init__(self, name):
        self.course_id = Course.id_counter  # Fixed: Course.id_counter instead of Course_id_counter
        Course.id_counter += 1
        self.name = name
        self.enrolled_students = []

    def __str__(self):
        return f"Course ID: {self.course_id}, Name: {self.name}, Enrolled: {len(self.enrolled_students)}"

    def enroll_student(self, student):
        if student not in self.enrolled_students:
            self.enrolled_students.append(student)
            # print(f"student (Student.name) enrolled success in (Student.enroll_in_course)")
            print(f"Student enrolled successfully in {self.name}")
        else:
            print(f"Student is already enrolled in {self.name}")
```

```python
c1 = Course("ML")
c1.enroll_student("Mohamed abdeen")
print(c1)
```
