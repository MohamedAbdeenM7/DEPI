```python
# Task 1
# Recursion
def fac_fun(num):

    if num == 0:
        return 1
    # if num == 1:
    #     return 1
    return num * fac_fun(num - 1)
fac_fun(5)
```

```python
# Task 2
# Prime Numbers
def is_prime(num):
    for i in range(2,num):
        if num % i == 0:
            return False
    return True
print(is_prime(3))
print(is_prime(12))
```

```python
# Task 3
# Common Dividors 

def dividors(num1:int, num2:int):

    mn = min(num1,num2) # ? mn + 1 ?
    common_div = []
    for i in range(1,mn + 1):
        if num1 % i == 0 and num2 % i == 0 :
            common_div.append(i)
    return common_div
dividors(10,20)
```

```python
# Task 4
# Generating Password 
import random
def gen_pas(length:int):

    char = "abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ1234567890!@#$%^&*"
    password = ""
    for _ in range(length): # _ not intersted in i but what it represent 
        password += random.choice(char)

    return password
gen_pas(5)
```

```python
# Task 5
# Extract the Unique Words
words = ["Welcome", "Ali", "Hi", "Ali", "No", "Hi", "No", "Ali", "No", "Ali"]

def unique_words(words):
    unique_word = set(words)
    counts  = {word: 0 for word in unique_word}
    for i in unique_word:  # o(n)
        for j in words:   #o(n)
            if i in  j :
                counts[i] +=1
    return counts,unique_word

unique_words(words)
```
