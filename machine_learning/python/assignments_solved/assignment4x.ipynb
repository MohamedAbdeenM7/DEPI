#### Task 1 – <b>(Simple Calculator Program)</b>
Objective:<br>
Develop a Python program that acts as a simple calculator. The program should allow the user to<br>
choose an arithmetic operation and apply it to two numbers entered by the user.<br>

```python
def add(n1,n2):
    return n1 + n2
def sub(n1,n2):
    return n1 - n2
def mul(n1,n2):
    return n1 * n2
def div(n1,n2):
    return n1 / n2
def main():
    print("Simple Calculator")
    operation = input("Please Enter The Desierd Operation:( +, -, *, / )")
    n1 = input("Please Entere The First Number:")
    n2 = input("Please Entere The Second Number:")
    if operation   == "+":
        return add
    elif operation == "-":
        return sub
    elif operation == "*":
        return mul
    elif operation == "/":
        return div
    else:
        print("error")
```

```python
main()
```
