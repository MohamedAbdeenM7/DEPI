#### Task 1 – <b>(Simple Calculator Program)</b>
Objective:<br>
Develop a Python program that acts as a simple calculator. The program should allow the user to<br>
choose an arithmetic operation and apply it to two numbers entered by the user.<br>

```python
def add(n1,n2):
    return n1 + n2
def sub(n1,n2):
    return n1 - n2
def mul(n1,n2):
    return n1 * n2
def div(n1,n2):
    return n1 / n2

def calc():
    print("Simple Calculator")
    n1 = float(input("Please Entere The First Number:"))
    operation = input("Please Enter The Desierd Operation:( +, -, *, / )")
    n2 = float(input("Please Entere The Second Number:"))

    if operation   == "+":
        result = add(n1,n2)
    elif operation == "-":
        result = sub(n1,n2)
    elif operation == "*":
        result = mul(n1,n2)
    elif operation == "/":
        result = div(n1,n2)
    else:
        print("error")
    return print(n1,operation,n2,"= ",result)
```

```python
calc()
```

#### Task 2 – <b>(Thanos Project)</b>  
Objective:<br>
Create a Python function that performs a series of file and folder operations,<br>
inspired by the 
concept of “Thanos” randomly removing half of the files.<br>
• The final number of files is displayed correctly. <br>

```python
import os
import random


os.makedirs("thanos",exist_ok=True) # to create the folder

for i in range(10):
    open(f"thanos/file#{i}.txt","w") # to create the files as txt and "w" for write mode


items = os.listdir("thanos")
half_items = int(len(items))

choices = []
for i in range(half_items):
    choice = random.choice(items)
    choices.append(choice)
    if i == 4 :
        break
    print(choices)

os.open("thanos")
for file in choices:
    os.remove(file)
    print(file)
```
