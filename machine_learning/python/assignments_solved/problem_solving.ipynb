### Problem 1 <b>(Reversed Words)</b>
##### Description:

Complete the solution so that it reverses all of the words within the string passed in.<br>
Words are separated by exactly one space and there are no leading or trailing spaces.<br>

Example(Input --> Output):<br>
"The greatest victory is that which requires no battle" --> "battle no requires which that is victory greatest The"

```python
# Problem 1 Solution

sentance = ('The greatest victory is that which requires no battle')
words = list(sentance.split(" "))
re_words = words[::-1]
re_sentance = " "
for i in re_words:
   print(re_sentance + i)

```
