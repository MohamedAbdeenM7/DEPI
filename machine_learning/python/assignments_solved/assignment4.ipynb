#### Task 1 – <b>(Simple Calculator Program)</b> 
Objective:<br>
Develop a Python program that acts as a simple calculator. The program should allow the user to<br>
choose an arithmetic operation and apply it to two numbers entered by the user.<br>

Description<br>
The program is menu-driven and performs basic arithmetic operations using separate functions for<br>
each operation. The user selects an operation, enters two numbers, and the program displays the <br>
calculated result.

Requirements 
1. Define arithmetic functions <br>
Create separate functions for: <br>
o Addition <br>
o Subtraction <br>
o Division <br>
o Multiplication <br>
2. Display a menu <br>
o Show a menu that allows the user to choose one of the available operations. 
3. User input <br>
o Ask the user to enter: <br>
o The desired operation <br>
o Two numeric values <br>
4. Perform the calculation <br>
o Call the appropriate function based on the user’s choice. <br>
o Handle invalid operation selections.<br> 
5. Display the result <br>
o Print the result of the selected operation clearly. 
6. Main program execution <br>
o Use a main() function. <br>
o Ensure the program runs only <br> when executed directly. <br>
<b>Notes</b> <br>
• The program should use functions to <br> keep the code organized. <br>
• User input must be converted to <br> numerical values before calculations. <br>
• The output should clearly show the result.

#### Task 2 – <b>(Thanos Project)</b>  
Objective: 
Create a Python function that performs a series of file and folder operations,<br>
inspired by the 
concept of “Thanos” randomly removing half of the files.<br>

Requirements 
Write a function that does the following steps:

1. Open a folder 
o Access an existing folder on your system. 
2. Create files inside the folder 
o Generate a specified number of files inside the folder. 
3. Count the files 
o Check and display the total number of files in the folder. 
4. Delete half of the files randomly 
o Randomly select and delete half of the files inside the folder. 
5. Check the result 
o Recount and display the remaining number of files after deletion.<br>

<b>Notes</b> <br>
• File deletion must be random. <br>
• The function should print clear messages showing each step and its result. <br>
• Use Python standard libraries such as os and random. <br>

Expected Outcome 
After running the function:<br> 
• Files are created successfully.<br> 
• Half of them are randomly removed. <br>
• The final number of files is displayed correctly. <br>
