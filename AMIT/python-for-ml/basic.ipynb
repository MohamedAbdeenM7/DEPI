```python
# Quiz 2 : Create a simple calculator
num_1 = float(input("Enter num_1 "))
operation = input("Enter operation (+, -, *, /): ")
num_2 = float(input("Enter num_2 "))
if operation == "+":
    result = num_1 + num_2
elif operation == "-":
    result = num_1 - num_2
elif operation == "*":
    result = num_1 * num_2
elif operation == "/":
    result = num_1 / num_2
else:
    result = "you entered an incorrect operation"
print("The result is: ", result)
```

```python
# Quiz 3 grading system 
degree = float(input("enter the degree"))
if 0 < degree < 100 :
    if  90 <= degree <= 100 :
        print("A")
    elif 80 <= degree < 90 :
        print("B")
    elif 70 <= degree < 80 :
        print("C")
    elif 60 <= degree < 70 :
        print("D")
    elif 50 <= degree < 60 :
        print("E")
    elif       degree < 50 :
        print("F")
else:
    print("error, the value must be between 0 to 100")
```
